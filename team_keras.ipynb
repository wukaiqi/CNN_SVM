```python
import sys 
import os
import urllib
import urllib.request
import urllib.parse
import urllib.error
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
#定义数据增强参数
datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        vertical_flip=True,
        horizontal_flip=True,        
        fill_mode='nearest')
```

```python
path="C:/Users/10449/Desktop/88"  
flag=True 
#b = 0
for name in os.listdir(path): 
    a=os.path.splitext(name)
    print (a[0])
    img = load_img(os.path.join(path,name))  # 获取地址
    x = img_to_array(img)  # 把PIL图像转换成一个numpy数组，形状为(3, 150, 150)
    x = x.reshape((1,) + x.shape)  # 这是一个numpy数组，形状为 (1, 3, 150, 150)
    word = 'stone'
    i = 0
    for batch in datagen.flow(x, batch_size=1,
                          save_to_dir='C:/Users/10449/Desktop/data_svm/test/1', save_prefix='', save_format='jpeg'
                         ):
        print("新建一张,已有" + str(i) + word)
        i += 1
        if i >9 :
            break  # 否则生成器会退出循环
print("end")
```
